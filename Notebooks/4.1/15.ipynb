```python
forward = lambda x, y, fx, fy: (fy - fx) / (y - x)
backward = lambda x, y, fx, fy: (fx - fy) / (x - y)
```

```python
xa = [0.5, .6, .7]
fxa = [.4794, .5646, .6442]
```

```python
def getVals(x, fx):
    fd, bd = [], []
    for i in range(0,2):
        fd.append(str(forward(x[i], x[i+1], fx[i], fx[i+1]))[:5])
        bd.append(str(backward(x[i], x[i+1], fx[i], fx[i+1]))[:5])
    return fd, bd
```

```python
ans1a = getVals(xa, fxa)
ans1a
```
